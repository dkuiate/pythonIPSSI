```python
from selenium import webdriver

browser = webdriver.Chrome()
```

```python
browser.get("https://www.leboncoin.fr/voitures/offres/ile_de_france/paris/p-2")
```

```python
voitures = browser.find_elements_by_css_selector(".Msj0v")
```

```python
for v in voitures:
    print(v.text)
```

```python
voitures = browser.find_elements_by_css_selector('._2pXp-')
for v in voitures:
    try:
        title = v.find_element_by_css_selector("p[data-qa-id='aditem_title']")
        price = v.find_element_by_css_selector("span[data-qa-id='aditem_price']")
        img = v.find_element_by_css_selector("img")
        print(f"Title : {title.text}")
        print(f"Price : {price.text}")
        print(f"img : {img.get_attribute('src')}")
    except:
        print("Error")

#     print(v.text)
    
```
